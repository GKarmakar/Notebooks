```python
#set seed for reproductability
import numpy as np
np.random.seed(42)
```

#### Load dependancies

```python
import keras
from keras.utils import np_utils
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
X_train = X_train.reshape(60000, 784).astype("float32")
X_test = X_test.reshape(10000, 784).astype("float32")
```

```python
X_train/= 255.
X_test/= 255.
```

```python
y_test.shape
```

```python
y_test[:10]
```

```python
n_classes = 10
y_train = np_utils.to_categorical(y_train, num_classes=n_classes)
y_test = np_utils.to_categorical(y_test, num_classes=n_classes)
```

```python
y_train.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
model = Sequential()
model.add(Dense(64, activation = "sigmoid", input_shape=(784,)))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(loss = 'mean_squared_error', optimizer=SGD(lr=0.01), metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=64, epochs=10, validation_data=(X_test, y_test))
```

```python
model = Sequential()
model.add(Dense(64, activation = "relu", input_shape=(784,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer=SGD(lr=0.01), metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=64, epochs=100, verbose = 1, validation_data=(X_test, y_test))
```
